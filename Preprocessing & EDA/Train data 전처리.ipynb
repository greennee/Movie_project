```python
# 감독데이터에 없음과 개봉연도없음을 숫자 0으로 바꾸기

df["감독"].replace(["없음", "개봉연도없음"], 0, inplace = True)
```

```python
# 감독 1명만 남기기

count = df["감독"].value_counts()
movies = []

for key, value in count.items():
    if value == 1:
        movies.append(key)
        
for i in range(len(df)):
    if df.iloc[i]["감독"] in movies:
        df["감독"].replace(df.iloc[i]["감독"], 0, inplace=True)
```

## 9. 개봉일

+ 개봉일이 없는 데이터는 1900-01-01로 만들기
+ 월과 일이 누락된 데이터는 -01 붙이기

```python
pd.options.display.float_format = '{:.5f}'.format
```

```python
# 결측인 데이터에 1900-01-01로 만들기

df['개봉일'].fillna("1900-01-01", inplace = True)
```

```python
# 개봉일 공백 제거

for i in range(len(df["개봉일"])):
    if type(df["개봉일"][i]) == float:
        df["개봉일"][i] = 0
    else:
        List = df["개봉일"][i].split(" ")
        df["개봉일"][i] = "".join(List)
```

```python
# 날짜 형식이 0000-00-00이 아닌 행의 인덱스 가져오기

index=[]

for i in range(len(df["개봉일"])):
    p= re.compile("\d\d\d\d-\d\d-\d\d")
    m= bool(p.match(str(df["개봉일"][i])))
    if m == False:
            index.append(i)
```

```python
# 날짜형식이 yyyy-mm-dd아닌 데이터 수정

for j in range(len(df['개봉일'])):
    if len(df.iloc[j]['개봉일']) == 4:
        df['개봉일'][j] = df.iloc[j]['개봉일'] + "-01-01"

    elif len(df.iloc[j]['개봉일']) == 7:
        df['개봉일'][j] = df.iloc[j]['개봉일'] + "-01"
        
        
    elif  len(df.iloc[j]['개봉일']) >= 11:
        df['개봉일'][j] = df.iloc[j]['개봉일'][-10:]
        
    else: 
        continue
```

```python
# 개봉년도 21년 이후 데이터 삭제

df = df[df["개봉일"] < "2021-01-01"]
```

```python
df.reset_index(inplace=True)
df.drop("index", axis=1, inplace = True)
```

## 10. 감독 / 배급사 / 배우 숫자화

- 해당 영화 개봉 6년 이내에 영화들에 대하여
    - 감독 : 6년동안 연출한 영화들의 누적관객수 합으로 치환
    - 배급사 : 6년동안 배급한 영화들의 누적관객수 합으로 치환
    - 배우 : 6년동안 출연한 영화들의 누적관객수 평균으로 치환
    - 배우들 : 배우는 최대 3명까지 포함되므로 배우들의 합으로 치환
    
- 감독 배급사 배우의 값이 모두 들어있는 원본데이터를 DB 데이터로 만들고 비교

```python
from datetime import datetime
from dateutil.relativedelta import relativedelta
import progressbar
```

```python
# 문자형으로 들어있는 개봉일을 날짜형으로 변경

db = pd.read_csv('DB_data.csv', encoding = 'utf-8')

df["개봉일"]=  pd.to_datetime(df["개봉일"], format="%Y-%m-%d")
db["개봉일"]=  pd.to_datetime(db["개봉일"], format="%Y-%m-%d")
```

```python
# 배우 -> 관객수 데이터로 치환

start = time.time()
bar = progressbar.ProgressBar()
df[["배우1","배우2","배우3"]] = 0

for nu in bar(range(len(df["배우"]))):
    if str(df["배우"][nu]).strip("[]").split(",")[0].strip("'") == 0:
        df["배우1"][nu] = 0 
        
    else:
        
        act_list = str(df.iloc[nu]["배우"]).strip("[]").split(", ")
        for numb in range(len(act_list)):
            count=0
            audi = 0
            for index in range(len(db["배우"])):  
                if act_list[numb].strip("'") in db["배우"][index] and (df.iloc[nu]["개봉일"] > db["개봉일"][index]) & (df.iloc[nu]["개봉일"]- relativedelta(years=6) < db["개봉일"][index]):
                        count+=1
                        audi += db["누적관객수"][index]
            if count != 0:
                audi = audi/count
                df["배우"+str(numb+1)][nu] = audi
                
end = time.time()
print(end - start) 
```

```python
# 감독 -> 관객수 데이터로 치환

start = time.time()
bar = progressbar.ProgressBar()
df["감독값"]= 0

for nu in bar(range(len(df["감독"]))):
    if df["감독"][nu] =='0':
        df["감독값"][nu] = 0 
        
    else:
        count=0
        audi = 0
        for index in range(len(db["감독"])):  
            if str(df["감독"][nu]) in db["감독"][index] and (df.iloc[nu]["개봉일"] > db["개봉일"][index]) & (df.iloc[nu]["개봉일"]- relativedelta(years=6) < db["개봉일"][index]):
                    count+=1
                    audi += db["누적관객수"][index]
        if count != 0:
            audi = audi/count
            df["감독값"][nu] = audi
            
end = time.time()
print(end - start) 
```

```python
# 배급사 -> 관객수 데이터로 치환

start = time.time()
bar = progressbar.ProgressBar()
df["배급사값"]= 0

for nu in bar(range(len(df["배급사"]))):
    if df["배급사"][nu] == 0:
        continue
        
    else:
        count=0
        audi = 0
        for index in range(len(db["배급사"])):  
            if df["배급사"][nu] in db["배급사"][index] and (df.iloc[nu]["개봉일"] > db["개봉일"][index]) & (df.iloc[nu]["개봉일"]- relativedelta(years=6) < db["개봉일"][index]):
                    count+=1
                    audi += db["누적관객수"][index]
        if count != 0:
            audi = audi/count
            df["배급사값"][nu] = audi
            
end = time.time()
print(end - start) 
```
