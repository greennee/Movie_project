## 배우 데이터 합, 평균 열 추가

```python
df["배우합"]= 0
df["배우평균"]=0

for i in range(len(df)):
    if len(df["배우"][i].strip("[]").split(", "))== 3:
        df["배우평균"][i] = (df.iloc[i]["배우1"]+ df.iloc[i]["배우2"] +df.iloc[i]["배우3"])/3
        df["배우합"][i] = df.iloc[i]["배우1"]+ df.iloc[i]["배우2"] +df.iloc[i]["배우3"]

    elif len(df["배우"][i].strip("[]").split(", "))== 2:
        df["배우평균"][i] = (df.iloc[i]["배우1"]+ df.iloc[i]["배우2"] + df.iloc[i]["배우3"])/2
        df["배우합"][i] = df["배우1"][i]+ df["배우2"][i]

    else:
        df["배우평균"][i] = df["배우1"][i]
        df["배우합"][i] = df["배우1"][i]
```
